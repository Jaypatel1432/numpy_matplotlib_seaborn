# 21. Question: Perform matrix multiplication of two 2D arrays using np.dot () and @. Compare the results and performance of both methods using a large dataset. 


```python
import numpy as np

import time

N = 1000  

A = np.random.rand(N, N)

B = np.random.rand(N, N)

start_dot = time.time()

result_dot = np.dot(A, B)

end_dot = time.time()

start_matmul = time.time()

result_matmul = A @ B

end_matmul = time.time()

print("Results are equal:", np.array_equal(result_dot, result_matmul))

print(f"Time taken with np.dot: {end_dot - start_dot} seconds")

print(f"Time taken with @: {end_matmul - start_matmul} seconds")
```
